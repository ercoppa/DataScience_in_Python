# Graphs and Network Analysis using NetworkX in Python

Graphs provide a powerful framework for modeling and analyzing relationships between objects. In this notebook, we will explore how to create and analyze graphs using the NetworkX library in Python. We will cover topics such as creating a graph, working with incidence and adjacency matrices, analyzing degrees, handling directed and undirected graphs, using real datasets, and examining various graph properties.

```python
import matplotlib.pyplot as plt
import pandas as pd

import networkx as nx
```

## Creating a Graph

In NetworkX, a graph can be created using the Graph class. Let's start by creating a simple undirected graph:

```python
G = nx.Graph()
print(G)
```

We can add nodes to the graph using the add_nodes_from method and edges using the add_edges_from method:

```python
G.add_nodes_from([1, 2, 3, 4, 5])
G.add_edges_from([(1, 2), (1, 3), (2, 4), (3, 4), (4, 5)])
```

## Plotting a Graph

NetworkX provides basic functionality for visualizing graphs, but its main goal is to enable graph analysis rather than perform graph visualization. We will use the matplotlib library for graph visualization. Let's plot the graph we created above:

```python
# plot
nx.draw(G, with_labels=True)
plt.show()
```

## Incidence Matrix
The incidence matrix is a representation of a graph as a binary matrix, where rows represent nodes and columns represent edges. 

The entry at row i and column j is 1 if node i is incident on edge j, and 0 otherwise. 

In NetworkX, we can obtain the incidence matrix using the incidence_matrix function:

```python
incidence_matrix = nx.incidence_matrix(G)
print("Incidence Matrix:")
print(incidence_matrix.toarray())
```

## Adjacency Matrix
The adjacency matrix is another common representation of a graph as a matrix. 

It is a square matrix where rows and columns represent nodes, and the entry at row i and column j is 1 if there is an edge between nodes i and j, and 0 otherwise. 

In NetworkX, we can compute the adjacency matrix using the adjacency_matrix function:

```python
adjacency_matrix = nx.adjacency_matrix(G)
print("\nAdjacency Matrix:")
print(adjacency_matrix.toarray())
```

## Degree
The degree of a node in a graph is the number of edges incident on it. It provides information about the connectivity of a node within the graph. In NetworkX, we can obtain the degrees of nodes using the degree function:

```python
degrees = dict(G.degree())
print("\nNode Degrees:")
print(degrees)
```

## Directed Graph and Undirected Graph
Graphs can be either directed or undirected. In a directed graph, edges have a specific direction, while in an undirected graph, edges do not have any direction. In NetworkX, we can create a directed graph using the DiGraph class and convert it to an undirected graph using the to_undirected method:

```python
D = nx.DiGraph()
D.add_edges_from([(1, 2), (2, 3), (3, 1)])
print(D)

U = D.to_undirected()
print(U)
```

## Real Example with Dataset

Karate Club Dataset is a social network of a karate club, where each node represents a member of the club and each edge represents a tie between two members of the club. The dataset captures the interactions between members of the club over a period of time. The dataset is provided by Zachary in 1977, and it is one of the most popular datasets used in social network analysis. The dataset is provided by NetworkX, and we can load it as follows:

```python
G = nx.karate_club_graph()
print(G)
```

```python
nx.draw(G, with_labels=True)
plt.show()
```

**Triadic Closure**: Triadic closure refers to the tendency of nodes that have a common neighbor to form connections between themselves. The triadic closure coefficient measures the extent to which triadic closure is present in the graph.

```python
triadic_closure = nx.transitivity(G)
print("\nTriadic Closure:")
print(triadic_closure)
```

**Clustering Coefficient**: The clustering coefficient measures the degree to which nodes in a graph tend to cluster together. It provides insight into the presence of clusters or communities within the graph.

```python
clustering_coefficient = nx.average_clustering(G)
print("\nClustering Coefficient:")
print(clustering_coefficient)
```

**Bridge**: A bridge in a graph is an edge whose removal increases the number of connected components. NetworkX provides the bridge_edges function to identify bridge edges in a graph. Let's find the bridge edges in our graph:

```python
bridge_edges = list(nx.bridges(G))
print("Bridge Edges:")
print(bridge_edges)
```

**Distance**: The distance between two nodes in a graph is the minimum number of edges required to travel from one node to another. NetworkX provides the shortest_path_length function to calculate the shortest path length between two nodes. Let's find the distance between nodes 1 and 5:

```python
distance = nx.shortest_path_length(G, source=1, target=5)
print("Distance between nodes 1 and 5:", distance)
```

**Diameter**: The diameter of a graph is the maximum distance between any pair of nodes in the graph. NetworkX provides the diameter function to calculate the diameter of a connected graph. Let's find the diameter of our graph:

```python
diameter = nx.diameter(G)
print("Graph Diameter:", diameter)
```

# Gephi
Gephi is an open-source software for visualizing and exploring graphs. NetworkX provides functionality to export graphs to Gephi's GEXF format. To export our graph to Gephi, we can use the write_gexf function:

```python
nx.write_gexf(G, 'graph.gexf')
```

You can use [Gephi](https://gephi.org/) to visualize the graph in a more interactive way.
